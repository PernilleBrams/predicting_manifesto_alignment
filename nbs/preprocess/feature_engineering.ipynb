## Adding government status
Determining the government status (coalition, opposition, support) of a party based on the date and name.

Data retrieved from https://www.ft.dk/da/folkestyret/regeringen/regeringer-siden-1953#EED8A1A7290F4732BE6E5CD19A199460 and https://folkevalgte.dk and https://www.regeringen.dk and https://danmarkshistorien.dk.

Government periods and their respective parties:
* 1994-1998: Two govs here, one without election: 
         1) (not relevant, not included in date range of data) 1994-1996: Poul Nyrup Gov II. Socialdemokratiet and Radikale Venstre and Centrum-Demokraterne (not included in data) in coalition, with Enhedslisten and Socialistisk Folkeparti supporting (ref: https://danmarkshistorien.dk/vis/materiale/folketingsvalget-1994/)
         2) 1996-1998: Poul Nyrup Gov III. Socialdemokratiet and Radikale Venstre in coalition, with Enhedslisten and Socialistisk Folkeparti supporting (ref: https://danmarkshistorien.dk/vis/materiale/folketingsvalget-1998/)

* 1998-2001: Majority government (just barely, by one Faroese seat). Poul Nyrup Gov IIII. Socialdemokratiet and Radikale Venstre in coalition. (ref: https://danmarkshistorien.dk/vis/materiale/folketingsvalget-1998/)

* 2001-2005: Anders Fogh I. Venstre and Konservative Folkeparti in coalition, with Dansk Folkeparti supporting. (ref: https://danmarkshistorien.dk/vis/materiale/folketingsvalget-2001/)

* 2005-2007: Anders Fogh II. Venstre and Konservative Folkeparti in coalition, with Dansk Folkeparti supporting. (ref: https://danmarkshistorien.dk/vis/materiale/folketingsvalget-2005/)

* 2007-2011: Anders Fogh III. Venstre and Konservative Folkeparti in coalition, with Dansk Folkeparti supporting. (ref: https://danmarkshistorien.dk/vis/materiale/folketingsvalget-2007/)

* 2011-2015: Helle Thorning Schmidt I. Socialdemokratiet, Radikale Venstre, and Socialistisk Folkeparti in coalition, with Enhedslisten supporting. (ref: https://danmarkshistorien.dk/vis/materiale/folketingsvalget-2011/)

* 2015-2019: Lars Løkke Rasmussen II. Venstre in a minority government with support from parties Liberal Alliance, and Konservative Folkeparti, but not a formal coalition, but then turned into a coalition of Venstre, Liberal Alliance, Konservative Folkeparti and support from Dansk Folkeparti the 28th november 2016.

* 2019-2022: Mette Frederiksen I. Socialdemokratiet in a minority government with support from Det Radikale Venstre, Socialistisk Folkeparti and Enhedslisten, but not a formal coalition.


```python
import pandas as pd

# Function to determine government status
def classify_government_status(row):
    """
    Only making code that targets handling these parties: 
    ['Socialdemokratiet', 'Dansk Folkeparti', 'Det Radikale Venstre',
       'Enhedslisten', 'Konservative Folkeparti',
       'Socialistisk Folkeparti', 'Venstre', 'Liberal Alliance',
       'Alternativet', 'Nye Borgerlige']
       
    Code makes the variable government_status with values coalition, support, opposition, along with government_status_finegrained (not used), with values coalition_maj (for majority gov), coalition_min (for minority gov), supporting, opposition
    
    From https://www.ft.dk/da/folkestyret/regeringen/regeringer-siden-1953#EED8A1A7290F4732BE6E5CD19A199460 and https://folkevalgte.dk and https://www.regeringen.dk and https://danmarkshistorien.dk

    Government periods and their respective parties:
    - 1994-1998: 
        Two govs here, one without election: 
        1) (not relevant, not included in date range of data) 1994-1996: Poul Nyrup Gov II. Socialdemokratiet and Radikale Venstre and Centrum-Demokraterne (not included in data) in coalition, with Enhedslisten and Socialistisk Folkeparti supporting (ref: https://danmarkshistorien.dk/vis/materiale/folketingsvalget-1994/)
        2) 1996-1998: Poul Nyrup Gov III. Socialdemokratiet and Radikale Venstre in coalition, with Enhedslisten and Socialistisk Folkeparti supporting (ref: https://danmarkshistorien.dk/vis/materiale/folketingsvalget-1998/)
    - 1998-2001: Majority government (just barely, by one Faroese seat). Poul Nyrup Gov IV. Socialdemokratiet and Radikale Venstre in coalition. (ref: https://danmarkshistorien.dk/vis/materiale/folketingsvalget-1998/)
    - 2001-2005: Anders Fogh I. Venstre and Konservative Folkeparti in coalition, with Dansk Folkeparti supporting. (ref: https://danmarkshistorien.dk/vis/materiale/folketingsvalget-2001/)
    - 2005-2007: Anders Fogh II. Venstre and Konservative Folkeparti in coalition, with Dansk Folkeparti supporting. (ref: https://danmarkshistorien.dk/vis/materiale/folketingsvalget-2005/)
    - 2007-2011: Anders Fogh III. Venstre and Konservative Folkeparti in coalition, with Dansk Folkeparti supporting. When Anders Fogh was elected general secretary of Nato, Lars Løkke Rasmussen took over as Prime Minister. (ref: https://danmarkshistorien.dk/vis/materiale/folketingsvalget-2007/)
    - 2011-2015: Helle Thorning Schmidt I. Socialdemokratiet, Radikale Venstre, and Socialistisk Folkeparti in coalition, with Enhedslisten supporting. (ref: https://danmarkshistorien.dk/vis/materiale/folketingsvalget-2011/)
    - 2015-2019: Lars Løkke Rasmussen II. Venstre in a minority government with support from parties Liberal Alliance, and Konservative Folkeparti, but not a formal coalition, but then turned into a coalition of Venstre, Liberal Alliance, Konservative Folkeparti and support from Dansk Folkeparti the 28th November 2016.
    - 2019-2022: Mette Frederiksen I. Socialdemokratiet in a minority government with support from Det Radikale Venstre, Socialistisk Folkeparti and Enhedslisten, but not a formal coalition.

    Args:
    - row (pd.Series): Row of the df containing 'ds' and 'unique_id'.

    Returns:
    - tuple: (government_status, government_status_finegrained)
    """

    date = row['Date']
    party = row['Party_Name']

    if date.year < 1998:
        if party in ['Socialdemokratiet', 'Det Radikale Venstre']:
            return 'coalition', 'coalition_min'
        elif party in ['Enhedslisten', 'Socialistisk Folkeparti']:
            return 'support', 'support'
        else:
            return 'opposition', 'opposition'
    elif date.year < 2001:
        if party in ['Socialdemokratiet', 'Det Radikale Venstre']:
            return 'coalition', 'coalition_maj'
        else:
            return 'opposition', 'opposition'
    elif date.year < 2005:
        if party in ['Venstre', 'Konservative Folkeparti']:
            return 'coalition', 'coalition_min'
        elif party == 'Dansk Folkeparti':
            return 'support', 'support'
        else:
            return 'opposition', 'opposition'
    elif date.year < 2007:
        if party in ['Venstre', 'Konservative Folkeparti']:
            return 'coalition', 'coalition_min'
        elif party == 'Dansk Folkeparti':
            return 'support', 'support'
        else:
            return 'opposition', 'opposition'
    elif date.year < 2011:
        if party in ['Venstre', 'Konservative Folkeparti']:
            return 'coalition', 'coalition_min'
        elif party == 'Dansk Folkeparti':
            return 'support', 'support'
        else:
            return 'opposition', 'opposition'
    elif date.year < 2015:
        if party in ['Socialdemokratiet', 'Det Radikale Venstre', 'Socialistisk Folkeparti']:
            return 'coalition', 'coalition_min'
        elif party == 'Enhedslisten':
            return 'support', 'support'
        else:
            return 'opposition', 'opposition'
    elif date.year < 2019:
        if party in ['Venstre', 'Liberal Alliance', 'Konservative Folkeparti']:
            return 'coalition', 'coalition_min'
        elif party == 'Dansk Folkeparti':
            return 'support', 'support'
        else:
            return 'opposition', 'opposition'
    elif date.year <= 2022:
        if party == 'Socialdemokratiet':
            return 'coalition', 'coalition_min'
        elif party in ['Det Radikale Venstre', 'Socialistisk Folkeparti', 'Enhedslisten']:
            return 'support', 'support'
        else:
            return 'opposition', 'opposition'
    else:
        return 'unknown', 'unknown'

# Apply function to df
complete[['government_status', 'government_status_finegrained']] = complete.apply(classify_government_status, axis=1, result_type='expand')
print(complete[['government_status', 'government_status_finegrained']].value_counts())

incomplete[['government_status', 'government_status_finegrained']] = incomplete.apply(classify_government_status, axis=1, result_type='expand')
print(incomplete[['government_status', 'government_status_finegrained']].value_counts())
```

```python
# Get data on main dfs
incomplete['Date'] = pd.to_datetime(incomplete['Date'])

# Function to find the number of seats based on the closest past election
def find_seats(row, results_df):
    party = row['Party_Name']
    date = row['Date']
    # Filter results for the current party and past elections only
    filtered_results = results_df[(results_df['Party'] == party) & (results_df['Date'] <= date)]
    if not filtered_results.empty:
        # Get the most recent election result
        return filtered_results.iloc[-1]['Seats']
    return None

# Apply the function to each row in main df
incomplete['seats'] = incomplete.apply(lambda row: find_seats(row, results_df), axis=1)

# Do it for the complete as well
complete['Date'] = pd.to_datetime(complete['Date'])

# Apply the function to each row in main df
complete['seats'] = complete.apply(lambda row: find_seats(row, results_df), axis=1)

# Display some of the data to confirm
complete
```

## Engineering days until and since election

```python
# Extract year, month and day in week in its own col
import pandas as pd

election_dates = [
    pd.Timestamp('1994-09-21'),
    pd.Timestamp('1998-03-11'),
    pd.Timestamp('2001-11-20'),
    pd.Timestamp('2005-02-08'),
    pd.Timestamp('2007-11-13'),
    pd.Timestamp('2011-09-15'),
    pd.Timestamp('2015-06-18'),
    pd.Timestamp('2019-06-05'),
    pd.Timestamp('2022-11-01')
]

# Gonna do it for both dfs - first ensure they are sorted
# Ensure dfs are sorted by date
incomplete = incomplete.sort_values('Date')
complete = complete.sort_values('Date')

# Function to calculate time until the next election
def time_until_next_election(row, election_dates):
    future_elections = [election for election in election_dates if election > row['Date']]
    if future_elections:
        return (min(future_elections) - row['Date']).days
    return None  # In case there are no future elections

# Function to calculate time since the last election
def time_since_last_election(row, election_dates):
    past_elections = [election for election in election_dates if election < row['Date']]
    if past_elections:
        return (row['Date'] - max(past_elections)).days
    return None  # In case there are no past elections

# Apply the functions to each row in the dfs
incomplete['days_until_election'] = incomplete.apply(time_until_next_election, axis=1, election_dates=election_dates)
incomplete['days_since_election'] = incomplete.apply(time_since_last_election, axis=1, election_dates=election_dates)
incomplete['days_until_election_2'] = incomplete['days_until_election']**2 # squared
incomplete['days_since_election_2'] = incomplete['days_since_election']**2 # squared

# For the comlpete dates one:
complete['days_until_election'] = complete.apply(time_until_next_election, axis=1, election_dates=election_dates)
complete['days_since_election'] = complete.apply(time_since_last_election, axis=1, election_dates=election_dates)
complete['days_until_election_2'] = complete['days_until_election']**2 # squared
complete['days_since_election_2'] = complete['days_since_election']**2 # squared
```

## Adding variables from ParlGov

```python
# Parlgov vars
import pandas as pd

# Parlgov data from https://www.parlgov.org/data-info/
parlgov = pd.read_csv(f"{directory_path}/data/raw/added_variables/parlgov/party-Table 1.csv", sep = ";").drop(['country_name_short','party_name_english','party_name_ascii','chess','country_id','party_id','family_id','family_name_short','party_name_short'], axis=1)

parlgov = parlgov[parlgov['country_name'] == 'Denmark']

parlgov['party_name'] = parlgov['party_name'].replace('Enhedslisten – De Rød-Grønne', 'Enhedslisten')
parlgov['party_name'] = parlgov['party_name'].replace('Konservative', 'Konservative Folkeparti')
parlgov['party_name'] = parlgov['party_name'].replace('Ny-Liberal Alliance', 'Liberal Alliance')
parlgov['party_name'] = parlgov['party_name'].replace('Socialdemokraterne', 'Socialdemokratiet')

# Filter by parties we want to see data for
parties_to_include = [
    'Socialdemokratiet', 'Dansk Folkeparti', 'Det Radikale Venstre',
    'Enhedslisten', 'Konservative Folkeparti',
    'Socialistisk Folkeparti', 'Venstre', 'Liberal Alliance',
    'Alternativet', 'Nye Borgerlige'
]

parlgov = parlgov[parlgov['party_name'].isin(parties_to_include)]

# Take a look
parlgov
```

```python
# Merge parlgov on data on incomplete df
incomplete = incomplete.merge(parlgov, left_on='Party_Name', right_on='party_name', how='left').drop(['party_name'],axis=1)
incomplete

# Merge parlgov on data on complete df
complete = complete.merge(parlgov, left_on='Party_Name', right_on='party_name', how='left').drop(['party_name'],axis=1)
complete
```

## Adding some final variables
* block (political left- or right wing)

```python
# Define a dictionary to map party names to blocks (as derived from Danmarkshistorien.dk)
party_to_block = {
    'Alternativet': 'left',
    'Dansk Folkeparti': 'right',
    'Det Radikale Venstre': 'left',
    'Enhedslisten': 'left',
    'Konservative Folkeparti': 'right',
    'Liberal Alliance': 'right',
    'Nye Borgerlige': 'right',
    'Socialdemokratiet': 'left',
    'Socialistisk Folkeparti': 'left',
    'Venstre': 'right'
}

# Function to assign block
def assign_block(row):
    return party_to_block.get(row['Party_Name'], 'unknown')

incomplete['block'] = incomplete.apply(assign_block, axis=1)
complete['block'] = complete.apply(assign_block, axis=1)
```
